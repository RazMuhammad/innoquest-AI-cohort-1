## Content Based Recommender System

This method recommends items based on the features of the items themselves. It looks at the attributes or characteristics of the items (such as genre, description, or keywords)

```python
import numpy as np 
import pandas as pd
```

```python
movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
# Keeping important columns for recommendation
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
# handle genres

movies.iloc[0]['genres']
```

```python
#The ast (Abstract Syntax Trees) module is imported to safely evaluate expressions from strings. It helps convert a string 
#representation of a Python object (like a list or dictionary) into an actual Python object.

import ast #for converting str to list

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['genres']
```

```python
# handle keywords
movies.iloc[0]['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# handle cast
movies.iloc[0]['cast']
```

```python
# Here i am just keeping top 3 cast

def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head()
```

```python
# handle crew

movies.iloc[0]['crew']
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# handle overview (converting to list)

movies.iloc[0]['overview']
```

```python
#type(x)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(4)
```

```python
movies.iloc[0]['overview']
```

```python
movies.head()
```

```python
# now removing space like that 
'Anna Kendrick'
'AnnaKendrick'

def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

```python
# Concatinate all
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies.iloc[0]['tags']
```

```python
# droping those extra columns
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# Converting list to str
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
# Converting to lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []
    
    for i in text.split():
        T.append(ps.stem(i))
    
    return " ".join(T)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
#len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

```python
new_df[new_df['title'] == 'The Lego Movie']
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
similarity[744]
```

```python
def recommend(movie):
    # Step 1: Get the index of the movie in the DataFrame
    index = new_df[new_df['title'] == movie].index[0]
    
    # Step 2: Compute the similarity (distances) between the input movie and all other movies
    distances = sorted(
        list(enumerate(similarity[index])),   # Step 2a: Enumerate the similarity scores of the input movie
        reverse=True,                         # Step 2b: Reverse the order to sort in descending order (higher similarity comes first)
        key=lambda x: x[1]                    # Step 2c: Sort based on the similarity score (second element of each tuple)
    )
    
    # Step 3: Print the titles of the 5 most similar movies (excluding the input movie itself)
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man 2')
```

```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('/kaggle/input/movie-data/u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("/kaggle/input/movie-data/Movie_Id_Titles")
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
